```python
#histogram with matplotlib
plt.figure(figsize=(6,3))
plt.hist(dataset['age'],
         bins=50,
         color='indigo',
         edgecolor='black',
         histtype='bar',#histype='step/stepfilled
         #density=True,
         cumulative=False,
         rwidth=0.9,
         orientation='vertical',#orientation='hosizontal'
         alpha= 0.9)
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Insurance age graph')
plt.show()
```

```python
#histogram with seaborn
plt.figure(figsize=(6,3))
sns.histplot(data=dataset,
             x='bmi',
             bins=25,
             kde=True,
             color='indigo',
             fill=False,
             element='bars',##step,poly
             )
plt.show()
```
