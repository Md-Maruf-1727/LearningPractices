```python
import pandas as pd
import numpy as np
import seaborn as sns
import requests
from io import StringIO
import matplotlib.pyplot as plt
```

```python
url1 = "https://gist.githubusercontent.com/netj/8836201/raw/6f9306ad21398ea43cba4f7d537619d0e07d5ae3/iris.csv"
data1 = requests.get(url1)
dataset1 = StringIO(data1.text)
iris_data = pd.read_csv(dataset1)
iris_data
```

```python
mainiris_data = sns.load_dataset('iris')
mainiris_data
```

```python
#heatmap
iris_corr = mainiris_data.select_dtypes('number')
sns.heatmap(iris_corr.corr(), annot=True)
```

```python
#lineplot
x = [23,98,25,67]
y = [34,56,34,97]
plt.plot(x,y, "ro")
```

```python
x = [56,52,21,56]
y = [78,23,45,32]
plt.plot(x,y)
```

```python
x = [56,52,21,56]
y = [78,23,45,32]
plt.figure(figsize=(3,2))
plt.plot(x,y)
plt.xlabel("value of X")
plt.ylabel("value of Y")
plt.title("Value of x and y")
```

```python
x = [56,52,21,56]
y = [78,23,45,32]
plt.figure(figsize=(3,2))
plt.plot(x,y, marker="^", color="g")
plt.xlabel("value of X")
plt.ylabel("value of Y")
plt.title("Value of x and y")
```

```python
x = [56,52,21,56]
y = [78,23,45,32]
plt.figure(figsize=(3,3))
plt.plot(x,y, "^k", color="b")
plt.xlabel("value of X")
plt.ylabel("value of Y")
plt.title("Value of x and y")
```

```python
#lineplot
x = [56,52,21,56]
y = [78,23,45,32]
plt.figure(figsize=(3,2))
sns.lineplot(x=x,y=y, color="g")
plt.xlabel("value of X")
plt.ylabel("value of Y")
plt.title("Value of x and y")
```

```python
## scatter plot
x = np.linspace(0,10,50)
y = np.sin(x)
plt.figure(figsize=(4,3))
plt.scatter(y,x, color="g")
plt.xlabel("value of X")
plt.ylabel("value of Y")
plt.title("Value of x and y")
```

```python
plt.Figure(figsize=(3,3))
plt.hist(mainiris_data['sepal_length'], bins=40, color='white', edgecolor='black', histtype='stepfilled')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Iris sepal length')
plt.show()
```

```python
plt.Figure(figsize=(3,3))
plt.hist(mainiris_data['sepal_length'], bins=40, color='white', edgecolor='black', histtype='bar')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Iris sepal length')
plt.show()
```
