```python
plt.figure(figsize=(3,3))
weights = np.ones_like(mainiris_data['sepal_length']) / len(mainiris_data['sepal_length'])
plt.hist(mainiris_data['sepal_length'],weights =weights, bins=40, color='white', edgecolor='black', histtype='stepfilled')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Iris sepal length')
plt.show()
```

```python
plt.figure(figsize=(4,3))
weights = np.ones_like
plt.hist(mainiris_data['sepal_length'], bins=40, color='green', edgecolor='black', histtype='bar')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Iris sepal length')
plt.show()
```

```python
plt.figure(figsize=(4,3))
weights = np.ones_like
plt.hist(mainiris_data['sepal_length'], cumulative=False, bins=40, color='green', edgecolor='black', histtype='bar')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Iris sepal length')
```

```python
plt.figure(figsize=(4,3))
plt.hist(mainiris_data['sepal_width'], label='Sepal Label',rwidth=0.5,alpha=0.7, bins=30, color='black', edgecolor='pink', histtype='bar')
plt.xlabel('Value')
plt.ylabel('frequency')
plt.title('Iris sepal length')
plt.legend()#use for seeing label
plt.show()
plt.grid(True)
```

```python
plt.figure(figsize=(6,3))
plt.hist(mainiris_data['petal_length'],label='petal width',histtype='step',alpha=0.5,rwidth=0.6,edgecolor='teal',color='teal', bins=50)
plt.xlabel('value')
plt.ylabel('frequency')
plt.title('petal width')
plt.legend()
plt.show()
plt.grid()
```

```python
#histgram with seaborn
plt.figure(figsize=(6,3))
sns.histplot(mainiris_data['petal_width'],fill=False,element='poly',kde=True, bins=50, color='maroon')
plt.xlabel('value')
plt.ylabel('frequency')
plt.title('petal width')
```

```python
plt.figure(figsize=(6,3))
sns.histplot(data=mainiris_data, x='petal_width',hue='species', multiple='layer',kde=True, bins=50, color='maroon')
plt.xlabel('value')
plt.ylabel('frequency')
plt.title('petal width')
```

```python
plt.figure(figsize=(6,3))
sns.histplot(data=mainiris_data, x='petal_length',hue='species', multiple='stack',kde=True, bins=50, color='maroon')
plt.xlabel('value')
plt.ylabel('frequency')
plt.title('petal width')
```

```python
plt.figure(figsize=(6,3))
sns.histplot(data=mainiris_data, x='petal_length',hue='species', multiple='dodge',kde=True, bins=30, color='maroon')
plt.xlabel('value')
plt.ylabel('frequency')
plt.title('petal width')
```

```python
#box plot with matplotlib
plt.figure(figsize=(6,3))
plt.boxplot(mainiris_data['sepal_length'], 
            label=['Sepal length'], 
            notch=True,
            patch_artist=True,
            boxprops=dict(facecolor='lime',color='tomato'),
            medianprops=dict(color='black', linewidth=2),
            flierprops=dict(marker='o', color='green'),
            showmeans=True)
plt.xlabel('Data')
plt.ylabel('Value')
```

```python
plt.figure(figsize=(5,3))
plt.boxplot(mainiris_data['sepal_width'],
            label=['Sepal Width'],
            vert=False,
            patch_artist=True,
            boxprops=dict(facecolor='deepskyblue',color='gray'),
            medianprops=dict(color='green', linewidth=2),
            flierprops=dict(marker='>', color='orange'),
            showmeans=True)
plt.xlabel('Data')
plt.ylabel('Value')
```

```python
plt.figure(figsize=(5,3))
plt.boxplot([mainiris_data['petal_length'], 
            mainiris_data['petal_width']],
            label=['Petal Length', 'Petal width'],
            notch=True,
            patch_artist=True,
            boxprops=dict(facecolor='honeydew', color='black'),
            medianprops=dict(color='black',),
            showmeans=True)
plt.xlabel('Data')
plt.ylabel('Value')
plt.show()
```

```python
# boxplot width with seaborn
plt.figure(figsize=(4,6))
sns.boxplot(data=mainiris_data,x='species', y='sepal_length',
            hue='species',
            palette='Dark2',
            fliersize=9,
            width=0.3)
```

```python
df_melt = mainiris_data.melt(value_vars=['petal_length', 'petal_width'],
                             var_name='Feature',
                             value_name='Value')
plt.figure(figsize=(6,4))
sns.boxplot(data=df_melt, x='Feature', y='Value',
            palette='Pastel1',
            width=0.2)
plt.title('Box Plot of Multiple Features')
plt.show()
```

```python
df_melt2 = mainiris_data.melt(value_vars=['sepal_length', 'petal_width'],
                              value_name="Value",
                              var_name="Features")
sns.boxplot(data=df_melt2, x='Features', y='Value',
            palette='Set3',
            width=0.5)
plt.show()
```

```python
#boxplot with matplotlib
plt.figure(figsize=(6,3))
plt.boxplot([mainiris_data['sepal_width'], mainiris_data['petal_length']],
            label=['sepal_width', 'petal_lentgh'],
            vert=False,
            patch_artist=True,
            boxprops=dict(facecolor='purple', color='green'),
            medianprops=dict(color='white', linewidth=2),
            flierprops=dict(marker='*', color='orange'),
            showmeans=True)
plt.show()
```

```python
#Scatter plot with matplotlib
plt.figure(figsize=(6,3))
plt.scatter(mainiris_data['sepal_length'],
            mainiris_data['sepal_width'],
            color='teal',
            marker='*',
            s=100,
            alpha=0.9,
            edgecolors='yellow',
            linewidths=1)
plt.xlabel('Sepal length')
plt.ylabel('Sepal Width')
plt.title('Scatter plot')
plt.show()
```

```python
plt.figure(figsize=(6,3))
plt.scatter(mainiris_data['petal_length'],
            mainiris_data['petal_width'],
            color='indigo',
            marker='X',
            edgecolors='green',
            s=50,
            linewidths=1,
            alpha=1,)
plt.xlabel('Petal length')
plt.ylabel('Petal Width')
plt.title('Scatter plot')
plt.show()
```

```python
#scatter plot with seaborn
plt.figure(figsize=(6,3))
sns.scatterplot(data=mainiris_data,
                x='sepal_length',
                y='petal_length',
                color= 'gray',
                style='species',
                hue='species',
                size=100,
                palette='Set1',
                alpha=1,
                s=50,
                legend='auto')
plt.xlabel('Setal length')
plt.ylabel('petal length')
plt.title('Scatter plot')
plt.show()
```

```python
plt.figure(figsize=(6,3))
sns.scatterplot(data=mainiris_data,
                x='sepal_width',
                y='petal_width',
                hue='species',
                style='species',
                size=1,
                palette='Set1',
                s=20,
                legend=False)
plt.xlabel('Setal Width')
plt.ylabel('Petal Width')
plt.title('Scatter plot')
plt.show()
```
